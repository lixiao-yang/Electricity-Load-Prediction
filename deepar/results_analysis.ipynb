# DeepAR Results Analysis

This notebook mirrors the structure of `tft/results_analysis.ipynb`, but focuses on the three DeepAR model groups: `2_3`, `7`, and `1_11`.

By default it selects the best random-search trial for each group using the lowest `overall_wmape`, generates plots, and saves them under `images/deepar/<group_key>`.

```python
from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd().resolve()
if PROJECT_ROOT.name == "deepar":
    PROJECT_ROOT = PROJECT_ROOT.parent
elif not (PROJECT_ROOT / "deepar").exists() and (PROJECT_ROOT.parent / "deepar").exists():
    PROJECT_ROOT = PROJECT_ROOT.parent

if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

from deepar.analysis import build_combined_evaluation_summary_table, build_summary_tables, run_full_analysis
```

```python
IMAGE_DIR = PROJECT_ROOT / "images" / "deepar"

bundle = run_full_analysis(
    image_root=IMAGE_DIR,
    selection_metric="overall_wmape",
    random_seed=42,
    history_days=60,
    show=False,
)

overall_summary, period_summary = build_summary_tables(bundle)
evaluation_summary_table = build_combined_evaluation_summary_table(bundle)
```

```python
display(evaluation_summary_table)
display(overall_summary)
display(period_summary)
```

```python
for group_key, group_bundle in bundle.items():
    print(f"\n===== Group {group_key} =====")
    print("Selected prediction file:", group_bundle["selected_prediction_path"])
    print("Selected user:", group_bundle["selected_user"])
    display(group_bundle["evaluation_summary_table"])
    display(group_bundle["trial_summary"].head())
    display(group_bundle["metrics_overall"])
    display(group_bundle["metrics_by_period"])
    display(group_bundle["user_period_summary"])
    display(group_bundle["horizon_metrics"].head(12))
    display(group_bundle["cluster_period_error_summary"])
```
